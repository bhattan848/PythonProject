```python
from bs4 import BeautifulSoup
import requests
soup = BeautifulSoup('http://www.nepalstock.com/floorsheet', 'html.parser')
url = requests.get(soup)
source = url.text
```

```python
source_code= BeautifulSoup(source, "lxml") 
```

```python
tag = source_code.table
tags = tag.find_all('tr')
print(tags)
```

```python
del tags[0:2]
```

```python
data = []
for i in tags:
    x = i.text
    data.append(x)
data
```

```python
final_data = []
for j in data:
    final_data.append(j.split('\n'))
print(final_data)
```
