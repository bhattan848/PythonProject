```python
from bs4 import BeautifulSoup
import requests 
soup = BeautifulSoup("https://www.worldometers.info/coronavirus/",'html.parser')
url = requests.get(soup)
source = url.text
```

```python
source_code = BeautifulSoup(source,"lxml")
# source_code
```

```python
tag = source_code.table
tags = tag.find_all('tr')
#print(tags)
```

```python
#tags[0]
```

```python
data = []
for i in tags:
    x = i.text
    data.append(x)
#data
```

```python
final_data = []
for j in data:
    print(j.split('\n')[1:-1])
```

```python
final_data = []
for j in data:
    final_data.append(j.split('\n')[1:-1])
#print(final_data)
```

```python
import csv
f = open('project.csv','a')
x = csv.writer(f)
for i in final_data:
    x.writerow(i)
```
